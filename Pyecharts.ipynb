# 更具官网的例子学习地图的制作

```python
from pyecharts.charts import Geo # 导入坐标
from pyecharts.datasets import *
# Map中只会设计到省, 如果要展现城市的数据, 则需要使用Geo,才可以
geo = Geo()

# 使用中国地图
geo.add_schema()
# 添加坐标点
datas = {
      "阿城": [126.58, 45.32],
      "阿克苏": [80.19, 41.09]
    }

for city, value in datas.items():
    geo.add_coordinate(city, value[0], value[1])

# geo.render_notebook()

# 使用已经有的地图数据项 
MapI

```
